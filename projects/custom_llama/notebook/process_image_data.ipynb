## Create New Meta File for MSCOCO


#### For Training Set

```python
from modelzipper.tutils import *


RAW_META_FILE = "/zecheng2/svg/mscoco/mscoco_train.jsonl"
SVG_DIR = "/zecheng2/svg/mscoco/convert_svg_train"
NEW_META_FILE = "/zecheng2/svg/mscoco/mscoco_train_svg.jsonl"

meta_file = auto_read_data(RAW_META_FILE)
all_files = auto_read_dir(SVG_DIR)

meta_file_dict = convert_list_to_dict(meta_file, "image_path")
all_svg_name = [os.path.basename(x).split('.')[0] for x in all_files]

new_meta_file = []

with tqdm(total=len(meta_file_dict.keys())) as pbar:
    for k in meta_file_dict.keys():
        name_ = os.path.basename(k).split('.')[0]
        if name_ in all_svg_name:
            new_meta_file.append(
                {
                    "image_path": meta_file_dict[k]["image_path"],
                    "captions": meta_file_dict[k]["caption"],
                    "svg_path": os.path.join(SVG_DIR, name_ + ".svg")
                }
            )
        pbar.update(1)

auto_save_data(new_meta_file, NEW_META_FILE)
```
